# 포지션별 상관분석

```python
# 포지션에 맞는 데이터 불러오는 함수
def load_data(position):
  data = pd.read_csv(data_path + 'epl_{}_data.csv'.format(position))
  del data['Unnamed: 0'], data['이름'], data['시즌'], data['포지션'], data['클럽팀']
  return data

# 연봉과의 상관계수를 반환하는 함수
def corr(data, method):
  data_pearson = data.corr(method=method)
  result = data_pearson['시장가치'].sort_values()
  return result
```

## 🟡 GK

```python
GK = load_data('GK')
corr(GK, 'pearson')
```

🟡 골키퍼의 시장가치와의 연관성 순위
1. 연봉: 0.587417

3. 승점: 0.492186
4. 출장 시간: 0.401501
5. 90분당 실점: -0.337732
